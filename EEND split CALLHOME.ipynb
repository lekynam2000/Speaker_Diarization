**Script to generate reco2dur**

```python
import soundfile as sf
```

```python
def generate_reco2dur(dir):
  rec2dur = {}
  lines = [line.strip().split(" ") for line in open(f"{dir}/wav.scp")]
  for line in lines:
    f = sf.SoundFile(f"{callhome_dir}/wav/{line[0]}.wav") 
    dur = f.frames/f.samplerate
    rec2dur[line[0]] = dur
  print(f"Saving reco2dur to {dir} ... ")
  with open(f"{dir}/reco2dur","w+") as file:
    for recid in rec2dur.keys():
      file.write(f"{recid} {rec2dur[recid]}\n")
```
